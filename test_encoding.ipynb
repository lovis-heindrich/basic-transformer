```python
from src.model import Embedding
import seaborn as sns
from matplotlib import pyplot as plt

sns.set(rc={'figure.figsize':(15,6)})
sns.set(font_scale=1.5)
sns.set_theme()

%load_ext autoreload
%autoreload 2
```

```python
# Recreate visualization from https://jalammar.github.io/illustrated-transformer/
embedding = Embedding(100, 64, 10)
encoding = embedding.encoding_vector.numpy()
assert encoding.shape == (10, 64)

g = sns.heatmap(encoding, cmap="viridis")
g.set_xlabel("Embedding dim")
g.set_ylabel("Position")
g.set_title("Positional embedding visualization")
plt.show()
```
