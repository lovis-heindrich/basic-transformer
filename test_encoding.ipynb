## Positional encoding

This notebook visualizes the positional encoding of the transformer implementation.

```python
from src.model import Embedding
import seaborn as sns
from matplotlib import pyplot as plt
from src.model import TransformerConfig

sns.set(rc={'figure.figsize':(15,6)})
sns.set(font_scale=1.5)
sns.set_theme()

%load_ext autoreload
%autoreload 2
```

```python
# Recreate visualization from https://jalammar.github.io/illustrated-transformer/
config = TransformerConfig(vocab_size=100, max_input_length=10, embedding_size=64)
embedding = Embedding(config)
encoding = embedding.encoding_vector.numpy()
assert encoding.shape == (10, 64)

g = sns.heatmap(encoding, cmap="viridis")
g.set_xlabel("Embedding dim")
g.set_ylabel("Position")
g.set_title("Positional embedding visualization")
plt.show()
```
